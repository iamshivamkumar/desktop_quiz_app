```python
from tkinter import *
import openpyxl
import random
class Quiz:
    
    def __init__(self):
        
        wb_obj = openpyxl.load_workbook("Book.xlsx")
        self.question_data = wb_obj.active
        self.max_r=self.question_data.max_row
        self.max_c=self.question_data.max_column
        self.ans_question_options = []
        self.q_list = []
        for i in range(1,self.max_r+1):
            self.q_list.append(i)
        random.shuffle(self.q_list)
        
        self.selected_option = ''
        self.ans = ''
        self.ans_message = ''
        self.window=Tk()
        self.window.geometry("800x500")
        self.display = Frame(self.window)
        self.display.pack(fill = 'both',expand = 1)
        self.submit_button = Button()
        self.clear_button = Button()
        self.next_button = Button()
        self.radio_button = []

    def next_question(self):
        self.ans_question_options = []
        for i in range(1, self.max_c + 1):
            cell_obj = self.question_data.cell(row = self.q_list[0], column = i)
            if cell_obj.value != None:
                self.ans_question_options.append(cell_obj.value)
        self.q_list.pop(0)
            
    def question_window(self):
        self.next_question()
        self.ans = self.ans_question_options[0]
        question = self.ans_question_options[1]
        options=[]
        for i in range(2,len(self.ans_question_options)):
            options.append(self.ans_question_options[i])
            
        self.display.forget()
        self.display.destroy()
            
        self.display = Frame(self.window)
        self.display.pack(fill = 'both',expand = 1)
            
            
        self.next_button = Button(self.display,text="Next",font=("Arial",10),state='normal',height=2,width=7,command=self.nextq)
        self.next_button.pack(padx=10,pady=10,anchor='ne')
        
        if len(self.q_list)==0:
            self.next_button['state'] = 'disabled'
            
        Label(self.display,text=question,font=("Arial",10),padx=15,pady=15).pack(anchor='w')
        self.selected_option=StringVar()
        self.selected_option.set(None)
            
        self.radio_button=[]
        for i in range(len(options)):
            self.radio_button.insert(i,Radiobutton(self.display,font=("Arial",10),command=self.radio_btn_select,text= options[i],padx=15,variable=self.selected_option,value=options[i]))
            self.radio_button[i].pack(anchor='w',side=TOP)
        label_frame = Frame(self.display)
        label_frame.pack()
            
        buttonframe = Frame(self.display)
        buttonframe.pack(anchor='se',side=BOTTOM)
            
        self.clear_button = Button(buttonframe,text='clear',height=2,width=7,font=("Arial",10),state='disabled',command=self.clear)
        self.clear_button.pack(side=LEFT,padx=10,pady=10)
                
        self.submit_button = Button(buttonframe,text='submit',height=2,width=7,font=("Arial",10),state='disabled',command = lambda:self.submit(label_frame))
        self.submit_button.pack(side=LEFT,padx=10,pady=10)
            
        self.window.mainloop()
    
    def radio_btn_select(self):
        self.submit_button['state'] = 'normal'
        self.clear_button['state'] = 'normal'
        
    def submit(self,label_frame):
        answer = self.ans
        answer = answer.strip().lower()
        select = self.selected_option.get()
        select = select.strip().lower()
        if answer.__eq__(select):
            self.ans_message='Correct answer'
            for widgets in label_frame.winfo_children():
                widgets.destroy()
            answer_label = Label(label_frame,font=("Arial",10),text=self.ans_message,padx=15,pady=15)
            answer_label.pack(anchor='w',side=TOP)
        else:
            self.ans_message="Wrong answer.The correct answer is:"
            for widgets in label_frame.winfo_children():
                widgets.destroy()
            answer_label=Label(label_frame,font=("Arial",10),text=self.ans_message+self.ans,padx=15,pady=15)
            answer_label.pack(anchor='w',side=TOP)
        self.submit_button['state'] = 'disabled'
        self.clear_button['state'] = 'disabled'
        for i in self.radio_button:
            i['state'] = 'disabled'
    def clear(self):
        self.selected_option.set(None)
        self.submit_button['state'] = 'disabled'
        self.clear_button['state'] = 'disabled'
        
    def nextq(self):
        self.question_window()
q=Quiz()
q.question_window()            
```
